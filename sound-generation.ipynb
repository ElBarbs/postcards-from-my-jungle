```python
!pip install torchsde

import torch
import soundfile as sf
from google.colab import userdata
from diffusers import StableAudioPipeline
from IPython.display import Audio  # Import the Audio module for playback.

pipe = StableAudioPipeline.from_pretrained("stabilityai/stable-audio-open-1.0", torch_dtype=torch.float16, token=userdata.get('HF_TOKEN'))
pipe = pipe.to("cuda")

# Define the prompts.
prompt = "Insect-like electronic shrieks. Sudden pitch modulations. Alternates between aggressive and inquisitive tones."
negative_prompt = "low bitrate, compressed, muffled, distorted, clipping, static, glitchy"

# Set the seed for generator.
generator = torch.Generator("cuda")

# Run the generation.
audio = pipe(
    prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=400,
    audio_end_in_s=10.0,
    num_waveforms_per_prompt=3,
    generator=generator,
).audios

output = audio[0].T.float().cpu().numpy()
sf.write("output.wav", output, pipe.vae.sampling_rate)

# Play the audio in Colab.
Audio("output.wav")
```
